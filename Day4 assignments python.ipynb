Question 2 : Function to return factorial

```

def fact(num):
     if num == 1:
        return num
     else:
      return num * fact(num-1)
num = int(input("Enter the number"))
if num < 0 :
      print( " no factorial for negative numbers")
elif num == 0 :
      print("Factorial of 0 is 1")
print("The factorial of ",num,"is",fact(num))
```
